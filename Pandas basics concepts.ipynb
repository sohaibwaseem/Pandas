### Conditional Selection
An important feature of pandas is conditional selection using bracket notation, very similar to numpy:

```python
df
```

```python
df > 0 # return boolian
```

```python
df > 1
```

```python
df == 1
```

```python
df[df>0] # Put NaN where the value is less than 0
```

```python
df['W'] > 0
```

```python
df[df['W'] > 0] # The row in which Value is less than Zero has removed
```

```python
df[df['W']>0]['Y']
```

```python
df[df['W'] > 0][['X', 'Z']]
```

#### Find Two Conditions
Never use 'And' OR 'or', however use '&' for 'and' and '|' for 'or'

```python
df[(df['W']>0) & (df['Y']>1)]
```

```python
df[(df['W']>1) | (df['Y']>1)]
```

## Multi Index or Index Hierarchy
##### Index Level

```python
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
list(zip(outside,inside))
```

```python
hier_index
```

```python
df1 = pd.DataFrame(randn(6,4), index = hier_index, columns = ['A', 'B', 'C', 'D'])
df1
```

```python
# locate 1st Hirerchy
df1.loc['G1']
```

```python
df1.loc['G1'].loc[1]
```

```python
df1.index.names
```

```python
df1.index.names = ["Group", "Num"]
```

```python
df1
```

```python
df1.xs('G1') # Return cross-section from the Series/DataFrame.
```

```python
df1.xs(['G2', 1])
```

```python
df1.xs(1,level = 'Num')
```
